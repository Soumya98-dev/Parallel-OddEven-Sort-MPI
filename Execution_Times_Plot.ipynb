```python
import matplotlib.pyplot as plt

# Data from timing_output.txt
array_sizes = [65536, 1048576, 16777216]
modified_times = [0.013283, 0.116871, 9.196096]
standard_times = [0.013016, 0.225104, 4.613275]
sequential_times = [0.010000, 0.190000, 3.380000]

# Plot execution times
plt.figure(figsize=(10, 6))
plt.plot(array_sizes, modified_times, label="Modified Odd-Even Sort", marker="o")
plt.plot(array_sizes, standard_times, label="Standard Odd-Even Sort", marker="s")
plt.plot(array_sizes, sequential_times, label="Sequential Quicksort", marker="^")
plt.xscale("log")
plt.yscale("log")
plt.xlabel("Array Size")
plt.ylabel("Execution Time (seconds)")
plt.title("Execution Time Comparison")
plt.legend()
plt.grid(True, which="both", linestyle="--")
plt.savefig("execution_times.png")
plt.show()
```
